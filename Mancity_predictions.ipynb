```python
mancity["target"] = (mancity["Result"] == "W").astype("int")
```

```python
mancity["poss"] = mancity["poss"].astype("int")
```

```python
mancity["passes"] = mancity["passes"].astype("int")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = mancity[mancity["Date"] < '4-11-2023']
```

```python
test = mancity[mancity["Date"]>"4-11-2023"]
```

```python
predictors = ["poss","passes"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(mancity, test_size=0.2, random_state=42)
print(test.shape)
```

```python
test = test.dropna()  # Might remove all rows if NaN values exist
test = test[test['Date'] > "04-11-23"]  # Might remove all rows if condition filters everything out
```

```python
test[predictors] = test[predictors].apply(pd.to_numeric, errors='coerce')
```

```python
preds = rf.predict(test[predictors].values)
```

```python
error = accuracy_score(test["target"], preds)
```
